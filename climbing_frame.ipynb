```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
import pandas_datareader.data as web
import datetime as dt
```

```python
# Get all focused stocks
symbol = ["AAPL", "BTC-USD", "FB", "GOOG", "MSFT", "TSLA"]
start = dt.date(2015, 1, 1)
end = dt.date.today()

all_in_one_df = web.DataReader(symbol, "yahoo", start, end)
all_in_one_df
```

```python
all_in_one_df.info()
```

```python
cf.set_config_file(offline = True)
```

```python
cf.colors.scales()
```

```python
cf.getThemes()
```

```python
all_in_one_df.Close.iplot(fill=True, colorscale="rdgy", theme="ggplot", title="Focused Stocks", xTitle="Time", yTitle="Stock Price")
```

```python
all_in_one_df.loc["2022-01":, ("Close","BTC-USD")].iplot()
```

```python
all_in_one_df.Close[["BTC-USD", "GOOG"]].iplot(kind = "spread", fill = True, colorscale= "rdylbu", theme= "solar",
                             title= "BTC-USD vs. GOOG", xTitle= "Time", yTitle= "Stock Price")
```
