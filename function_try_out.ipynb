```python
for day in df_aapl_b.index[:-126]:
    df_aapl_b.halfyear[day] = (df_aapl_b.close[day+126] / df_aapl_b.close[day]) * 100
```

```python
for day in df_aapl_b.index[:-252]:
    df_aapl_b.year[day] = (df_aapl_b.close[day+252] / df_aapl_b.close[day]) * 100
```

```python
for day in df_aapl_b.index:
    df_aapl_b.total[day] = (df_aapl_b.close[df_aapl_b.index[-1]] / df_aapl_b.close[day]) * 100
```

```python
def revenue_sorted(df):
    df = df.resample("B").mean().dropna()
    df = df.reset_index()
    df['week'] = np.nan
    df['month'] = np.nan
    df['quartal'] = np.nan
    df['halfyear'] = np.nan
    df['year'] = np.nan
    df['total'] = np.nan
    for day in df.index[:-5]:
        df.week[day] = (df.close[day+5] / df.close[day]) * 100
    for day in df.index[:-22]:
        df.month[day] = (df.close[day+22] / df.close[day]) * 100
    for day in df.index[:-63]:
        df.quartal[day] = (df.close[day+63] / df.close[day]) * 100
    for day in df.index[:-126]:
        df.halfyear[day] = (df.close[day+126] / df.close[day]) * 100   
    for day in df.index[:-252]:
        df.year[day] = (df.close[day+252] / df.close[day]) * 100
    for day in df.index:
        df.total[day] = (df.close[df.index[-1]] / df.close[day]) * 100
    df = df.set_index('timestamp')
    return df
    
```
